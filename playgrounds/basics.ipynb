```python
# imports
import matplotlib.pyplot as plt
import numpy as np
```

## Generating Data

```python
N = 1000
a = 2
b = 10
epsilon = 50

rng = np.random.default_rng(seed=42)

X_train = np.linspace(0, 100, num=N)
y_train = b + X_train * a + rng.normal(0, epsilon, size=N)

coeffs = np.polyfit(X_train, y_train, deg=1)
y_pred = np.polyval(coeffs, X_train)
```

## Plotting

### Scatter, Plot

```python
plt.scatter(X_train, y_train)
plt.xlabel("X_train")
plt.ylabel("y_train")
plt.plot(X_train, y_pred, color="red")
plt.title("Linear regression with noisy data")
# add legend
plt.legend(["Predicted line", "Training data"])
plt.show()
```

### Fill

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

t = np.linspace(0, 1, 3)
x = np.concatenate([t, 1-t, -t, -1+t])
y = np.concatenate([1-t, -t, -1+t, t])
ax = axs[0]
ax.scatter(x, y, color='black')
ax.fill(x, y, alpha=0.3)

x = np.array([-1, -1,-1, 0, 1, 1, 1, 0])
y = np.array([-1, 0, 1, 1, 1, 0, -1, -1])
ax = axs[1]
ax.scatter(x, y, color='black')
ax.fill(x, y, alpha=0.3)
ax.annotate('$(1, 0)$\n$d=1$',
            (1, 1),
            textcoords='offset points', 
            xytext=(5, 5), 
            fontsize=9)


for ax in axs:
    ax.set_aspect('equal')
    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)
    ax.axhline(0, color='gray', linewidth=0.5)
    ax.axvline(0, color='gray', linewidth=0.5)
    ax.set_xlabel('$x_1$')
    ax.set_ylabel('$x_2$')
```
